Cleaning and exploring The Data

```python
### to see which columns have null values using heatmap
sns.heatmap(df_movies.isnull())
```

```python
###Cleaning all the Null values from both the Dataframe

df_clean_movies=df_movies.dropna(axis = 1)
```

```python
df_clean_tv_shows = df_tv_shows.dropna(axis =1)
```

```python
df_clean_movies.isnull().sum()
```

```python
df_clean_tv_shows.isnull().sum()
```

```python
### Joining both the dataframes

Netflix_df = pd.concat([df_clean_movies, df_clean_tv_shows], ignore_index=True)
```

```python
Netflix_df
```

```python
### Removing the null values and columns
Netflix_df = Netflix_df.dropna(axis =1)
```

```python
Netflix_df.isnull().sum()
```

```python
## Creating New columns "month" and "year"
Netflix_df["month"] = pd.to_datetime(Netflix_df['date_added']).dt.month
Netflix_df["year"] = pd.to_datetime(Netflix_df['date_added']).dt.year
#### Drop column "show_id"
Netflix_df = Netflix_df.drop("show_id", axis=1)
```

```python
### Types of Languages
Netflix_df["language"].unique()
```

```python
#   a function to apply to every row to create a column named "Country"
def country(row: pd.Series):
    # Use str() and handle potential NaN values
    lang_val = row.get("language")
    if pd.isna(lang_val):
        return "Global"

    language = str(lang_val).lower()

    if "en" in language:
        return "United States"
    elif "ko" in language:
        return "South Korea"
    elif "es" in language:
        return "Latin"
    elif "ja" in language:
        return "Japan"
    elif "hi" in language:
        return "India"
    elif "de" in language:
        return "Germany"
    elif "ar" in language:
        return "Middle East"
    elif "zh" in language:
        return "Asian"
    else:
        return "Global"
```

```python
### the "country" column has been added by applying the function
Netflix_df["country"] = Netflix_df.apply(country, axis=1)
```

```python
Netflix_df = Netflix_df.set_index("date_added")
```

Plot and Statistical Analysis of the Data

```python
import seaborn as sns

sns.set_theme()

ax = sns.lineplot(data=Netflix_df, x="year", y="popularity", hue="type")

ax.set(
    title="Netflix Popularity Trends Over Time",
    xlabel="Year",
    ylabel="Average Popularity Score"
)
```

This line plot illustrates the strategic transformation of Netflix over the last 15 years:

*   **Strategic Pivot:** The data shows Netflix's shift from a **Movie-heavy library** (2010–2015) to a **TV-driven powerhouse** (2018–2025).
*   **The Engagement Gap:** While the catalog contains a high volume of movies, the **TV Show line** displays steeper growth, proving that series are the primary driver of global "buzz" and subscriber retention.
*   **Modern Momentum:** The sharp upward trend leading into **2025** highlights the success of Netflix's "Global Franchise" strategy, where international hits now dominate cultural conversation.

```python
old_df = Netflix_df[Netflix_df["year"] <= 2017]
sns.barplot(data=old_df, x="year", y="popularity", hue="type")
```

This bar plot captures the critical years when Netflix shifted its business model from a "digital library" to a "production powerhouse."

Strategic Pivot (2013–2015):** Notice the climb in **TV Show** popularity starting around 2013. This marks the debut of the first *Netflix Originals*, proving that exclusive series could drive more "buzz" than licensed movies.
The Content Mix:** In the early 2010s, popularity scores were relatively low and balanced. As we move toward 2017, the height of the bars increases, reflecting Netflix's successful **global expansion**.
 By the end of this era (2017), TV Shows had solidified their place as the primary driver of platform engagement, setting the stage for the "Global Explosion" seen in the next decade.

```python
new_df = Netflix_df[Netflix_df["year"] >= 2017]
sns.barplot(data=new_df, x="year", y="popularity", hue="type")
```

This plot illustrates the "modern" stage of Netflix, where the platform evolved into a global cultural phenomenon dominated by massive viral hits.

The Popularity Surge:** Notice that the Y-axis (Popularity) reaches much higher levels than the previous era. This represents the "viral effect"—where global hits like *Squid Game* or *Stranger Things* dominate worldwide conversation.
Dominance of TV Shows:** In almost every year of this era, **TV Shows** (orange) significantly outperform Movies (purple) in average popularity. This confirms that serialized content is the primary engine for high-engagement "binge-watching."
The 2025 Momentum:** The bars for **2024 and 2025** capture Netflix's current pivot into live events and massive franchise sequels, showing that global reach is currently at an all-time peak.

#Linear Regression

```python
import statsmodels.api as sm
import numpy as np

# Let's run an OLS regression to determine the relationship between popularity  and year

y = Netflix_df["popularity"].to_numpy()
X = sm.add_constant(Netflix_df["year"].to_numpy())
model = sm.OLS(y, X).fit()

X_predict = np.array([1, 2026])  # The 1 here is the constant, needed for the regression model.
predictions = model.predict(X_predict)
predictions
```

```python
model.summary()
```
